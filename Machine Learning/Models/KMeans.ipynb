## Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 

sns.set_theme()

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

## Loading the dataset

```python
df = pd.read_csv("nasa.csv")
df
```

```python
kmeans = KMeans(n_clusters=2).fit(df[["Absolute Magnitude", "Orbital Period"]])#.values.reshape(-1,1))
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
scores = [KMeans(k).fit(df[["Absolute Magnitude", "Orbital Period"]]).score(df[["Absolute Magnitude", "Orbital Period"]]) for k in range(1, 10)]
plt.plot(range(1, 10), scores, markersize=3, marker="o")
plt.show()
```
