# Attrition = No

```python
df_No = df.loc[df['Attrition']=="No"]
df_No
```

```python
dfhours = df_No[df_No.columns[11:13]]
dfhours

dfhours.describe()
```

```python
#EnvironmentSatisfaction
elow = 0
ehigh = 0

n_No = len(df_No[df_No.columns[10:11]])

e = df_No[df_No.columns[10:11]].values

for i in e:
    if 1 <= i <= 2:
        elow += 1
    if 3 <= i <= 4:
        ehigh += 1
    

pelow = round(elow/n_No * 100, 2)
pehigh = round(ehigh/n_No * 100, 2)
        
print("Number of employees with low Environment Satisfaction: " + str(elow) + " which is " + str(pelow) + " %.")
print("Number of employees with high Environment Satisfaction: " + str(ehigh) + " which is " + str(pehigh) + " %.")
```

```python
#JobInvolvement
jilow = 0
jihigh = 0

ji = df_No[df_No.columns[13:14]].values

for i in ji:
    if 1 <= i <= 2:
        jilow += 1
    if 3 <= i <= 4:
        jihigh += 1

pjilow = round(jilow/n_No * 100, 2)
pjihigh = round(jihigh/n_No * 100, 2)

print("Number of employees with low Job Involvement: " + str(jilow) + " which is " + str(pjilow) + " %.")
print("Number of employees with high Job Involvement: " + str(jihigh) + " which is " + str(pjihigh) + " %.")
```

```python
#Gender
female = 0
male = 0

gender = df_No[df_No.columns[11:12]].values

for i in gender:
    if i == "Female":
        female += 1
    if i == "Male":
        male += 1
        
pfemale = round(female/n_No * 100,2)
pmale = round(male/n_No * 100,2)
        
print("Number of Female employees: " + str(female) + " which is " + str(pfemale) + " %.")
print("Number of Male employees: " + str(male) + " which is " + str(pmale) + " %.")
```
